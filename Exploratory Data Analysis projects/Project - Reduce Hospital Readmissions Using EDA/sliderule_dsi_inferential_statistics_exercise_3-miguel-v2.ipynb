# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

#### B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

##### 1) Setup an appropriate hypothesis test.

Null hypothesis: "There is 0 correlation between the Number of Discharges and the Excess Readmission Ratio"

Alternate hypothesis: "There is a correlation between the Number of Discharges and the Excess Readmission Ratio"

test statistic: "pearson correlation coefficient"

##### 2) Compute and report the observed significance value (or p-value).

I will do permutations on the 'Number of Discharges' while leaving the 'Excess Readmission Ratio' fixed, since it is computationally more efficient. This will simulate the hypothesis that they are totally independent of each other.

```python
# pearson correlation coefficient formula
def pearson_r(data_1, data_2):
    return np.corrcoef(data_1, data_2)[0, 1]

# observed correlation coefficient
obs_corr = pearson_r(x, y)
print("The observed correlation coefficient is", obs_corr)


# 10000 permutations
perm_replicates = np.empty(10000)

for i in range(len(perm_replicates)):
    x_permuted = np.random.permutation(x)
    perm_replicates[i] = pearson_r(x_permuted, y)

# p-value
p_value = np.sum(perm_replicates <= obs_corr) / len(perm_replicates)

print("p-value:", p_value)
```

```python
# distribution of the permutations made
sns.distplot(perm_replicates)
plt.xlabel("correlation coefficient")
plt.show()
```

Admitting a significance level of 5%, we have statistical evidence to reject the null hypothesis that states that there is 0 correlation between the Number of Discharges and the Excess Readmission Ratio.

We are now going to see what other variables may influence the 'Excess Readmission Ratio'. Let us start with the procedures in the program:

I will first split the data between the two types of readmission ratio, so that we can focus on what can influence the classification of the readmission ration as an 'excess'. 

Then, have a visual cue of the impact of the variable 'Measure Name'.

```python
# spliting between excess and low readmission ratio
excess = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'] > 1]
low = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'] <= 1]
```

```python
# conditions/procedures
measure_names = clean_hospital_read_df['Measure Name'].unique()

# function for computing the proportion on the totality of the data
def prop_meas(meas_name, df):
    prop = df[df['Measure Name'] == meas_name]['Excess Readmission Ratio'].count() / len(clean_hospital_read_df)
    return prop

# bar plot
for i, j in enumerate(measure_names):
    p1 = plt.bar(i, prop_meas(j, excess), color='crimson')
    p2 = plt.bar(i+0.05, prop_meas(j, low), color='darkseagreen')
    
# formatting
plt.legend(('excess', 'low'), fontsize=16)
plt.ylabel("Percentage", fontsize=15)
plt.xlabel("Condition", fontsize=15)
plt.xticks([0,1,2,3,4], list(measure_names))
plt.title("Percentage of 'excess' and 'low' Readmission Ratio in the data", fontsize=16)
    
plt.subplots_adjust(left=0, bottom=-0.5, right=1.7, top=0.8, wspace=0.6, hspace=0.4)
plt.show()
```

We see that certain conditions, possibly the 'Acute Myocardial Infarction (AMI)' and 'Elective Primary Total Hip Arthroplasty and/or Total Knee Arthroplasty (HIP-KNEE)', could drive the excess readmission ratio. To ascertain if this could be the case, we will first hypothesize if the variable has an impact on the classification of the readmission ratio by running a Chi-squared test.

* **Null hypothesis**: "The variable measured procedure does not have an impact on the readmission ratio"
* **Alternate hypothesis**: "The variable measured procedure has an impact on the readmission ratio".

test statistic: "number of measured procedures"

```python
# function for counting
def count_meas(meas_name, df):
    count = df[df['Measure Name'] == meas_name]['Excess Readmission Ratio'].count()
    return count

# function for computing the proportion in the dataframe
def prop_meas_2(meas_name, df):
    prop = df[df['Measure Name'] == meas_name]['Excess Readmission Ratio'].count() / len(df)
    return prop

# observed and expected values
observed_values = []
expected_values = []

for i in measure_names:
    observed_values.append(count_meas(i, excess))
    expected_values.append(prop_meas_2(i, low))
                   
expected_values = [round(x * len(excess)) for x in expected_values]
        
print("Observed values:", observed_values)
print("Expected values:",expected_values)

print("Chi-squared test:", scipy.stats.chisquare(observed_values, expected_values))
```

Assuming a significance level of 5%, we have sufficient statistical evidence to reject the null hypothesis, which states that the measured procedure does not have an impact on the classification of the readmission ratio. Our initial suspicion was correct.

We can go further and say that if there are conditions which can affect this result, the two mentioned before are the most likely candidates, but we will test them all. If one of them does not have influence, then the observed difference between the number on excess and low readmission ratio could have been a chance event.

```python
# function for generating a permutation replicate
def permutation_sample(data1, data2):
    data = np.concatenate((data1, data2))
    permuted_data = np.random.permutation(data)
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]
    return perm_sample_1, perm_sample_2

# Generating permutation replicates
def draw_perm_reps(data_1, data_2, func, size=1):
    perm_replicates = np.empty(size)
    for i in range(size):
        perm_sample_1, perm_sample_2 = permutation_sample(data_1, data_2)
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)
    return perm_replicates

# computing difference between excess and low ratios
def diff_excess(data_1, data_2):
    nr_excess = sum(i > 1 for i in data_1) - sum(i > 1 for i in data_2)
    return nr_excess    
```

```python
# for each condition:

for i in measure_names:
    # dataframe
    df_condition = clean_hospital_read_df[clean_hospital_read_df['Measure Name'] == i]
    
    # spliting between excess and low readmission ratio
    df_condition_excess = df_condition[df_condition['Excess Readmission Ratio'] > 1]['Excess Readmission Ratio']
    df_condition_low = df_condition[df_condition['Excess Readmission Ratio'] <= 1]['Excess Readmission Ratio']
    
    # observed difference
    obs_diff = len(df_condition_excess) - len(df_condition_low)
    
    # simulations
    perm_replicates = draw_perm_reps(df_condition_excess, df_condition_low, diff_excess, size=1000)
    
    # p-value
    p_value = np.sum(abs(perm_replicates) >= abs(obs_diff)) / len(perm_replicates)
    
    print("probability value for", i, "=", p_value)
```

As we suspected, the p-values regarding the conditions 'Acute Myocardial Infarction (AMI)' and 'Elective Primary Total Hip Arthroplasty and/or Total Knee Arthroplasty (HIP-KNEE)' are below 1%, so we have statistical evidence to support the argument that these differences are significant, with a high degree of confidence. 

All other variables being equal, there is a greater incidence in excess readmission ratios when they are used to measure performance for those two conditions, which also means that a patient should not expect better hospital care in those cases. Hence, another recommendation could be to provide hospitals the necessary resources to deal with these conditions.
